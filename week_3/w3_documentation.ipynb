### 1. Identifying the Components

**1.1. What are the inputs?**

* Three individual quiz marks entered by the user.
* The user's choice (Y/N) to continue entering marks for another student.

**1.2. What is the process?**

* Calculate the average mark by adding the three quiz marks together and dividing by 3.
* Check if the calculated average is greater than or equal to 50 to determine the pass/fail status.
* Loop the entire process so another student's marks can be entered if requested.

**1.3. What is the output?**

* The student's calculated average mark.
* A message stating whether the student passes or fails.
* A final "Program Ended" message when the loop terminates.

### 2.2. Identify the Algorithm Structure

| Structure Type | What It Means | How It Applies to Your Code |
| :--- | :--- | :--- |
| **2.2.1. Sequence** | Executing a series of steps one after another in a specific order. | Asking the user for `quiz_1`, `quiz_2`, and `quiz_3` sequentially, and then calculating the `average` based on those inputs. |
| **2.2.2. Selection** | Making decisions and choosing different paths based on a specific condition. | The `if average >= 50:` statement that branches the program to display either a "Pass" or "Fail" result. |
| **2.2.3. Iteration** | Repeating a set of instructions until a certain condition is met (looping). | The `while choice == "y":` loop that encapsulates the main code, allowing you to enter marks for multiple students until the user types "N". |
